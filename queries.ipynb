```python
import pandas as pd
from pandasql import sqldf

# Charger le fichier CSV
df = pd.read_csv('./db/vgsales.csv')

# Fonction pour exécuter des requêtes SQL sur le DataFrame
pysqldf = lambda q: sqldf(q, globals())

# 1. Requête : Tous les jeux sur NES
nes_games = pysqldf("SELECT * FROM df WHERE Platform = 'NES'")
print("\nJeux sur NES (SQL) :")
print(nes_games)

# 2. Requête : Somme des ventes japonaises
japan_sales_sum = pysqldf("SELECT SUM(JP_Sales) as Total_JP_Sales FROM df")
print("\nSomme des ventes japonaises (SQL) :")
print(japan_sales_sum)

# 3. Requête : Nombre de jeux sortis après 2008
games_after_2008 = pysqldf("SELECT COUNT(*) as Games_After_2008 FROM df WHERE Year >= 2008")
print("\nNombre de jeux sortis après 2008 (SQL) :")
print(games_after_2008)

# 4. Requête : Pourcentage de jeux Wii
wii_percentage_query = """
SELECT 
    (CAST(SUM(CASE WHEN Platform = 'Wii' THEN 1 ELSE 0 END) AS FLOAT) / COUNT(*)) * 100 AS Wii_Percentage 
FROM df
"""
wii_percentage = pysqldf(wii_percentage_query)
print("\nPourcentage de jeux Wii (SQL) :")
print(wii_percentage)
```
